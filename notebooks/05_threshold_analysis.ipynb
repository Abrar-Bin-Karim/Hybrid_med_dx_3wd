```python
import pandas as pd
import matplotlib.pyplot as plt

sweep_df = pd.read_csv("../results/three_way_sweep.csv")

display(sweep_df.head())
```

```python
print(sweep_df.columns)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    sweep_df["coverage"],
    sweep_df["accuracy_non_uncertain"],
    alpha=0.5,
    label="Threshold Combinations"
)

chosen = sweep_df[
    (sweep_df["alpha"] == 0.91) &
    (sweep_df["beta"] == 0.19)
]

plt.scatter(
    chosen["coverage"],
    chosen["accuracy_non_uncertain"],
    color="red",
    s=120,
    label="Chosen Threshold (α=0.91, β=0.19)"
)

plt.xlabel("Coverage (Non-Uncertain Rate)")
plt.ylabel("Accuracy on Non-Uncertain Cases")
plt.title("Three-Way Decision Trade-Off Curve")
plt.legend()
plt.grid(True)

# SAVE BEFORE SHOW
plt.savefig("../results/threshold_tradeoff_plot.png", dpi=300, bbox_inches="tight")

plt.show()
```
